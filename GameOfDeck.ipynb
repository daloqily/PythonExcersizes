```python
from random import shuffle

##-----------------from resource-----------------------##
class Card:

  ## suits&value are represented by int



  def __init__(self, v,s):
      suits = ["spade ♠",
             "heart ♥",
             "diamonds ♦ ",
             "club ♣"]
    
      values = [None, None,"2", "3",
              "4", "5", "6", "7",
              "8", "9", "10",
              "Jack", "Queen",
              "King", "Ace"]
              
      self.value = values[v]
      self.suit = suits[s]

          ##  magic (dunder) functions that enables python to compare self defined objects
      ##less than##
  def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.suit < c2.suit:
                return True
            else:
                return False
        return False
      ##greater than##
  def __gt__(self, c2):
        if self.value > c2.value:
            return True
        if self.value == c2.value:
            if self.suit > c2.suit:
                return True
            else:
                return False
        return False

      ## magic methods that python interperets when printing a self-defined object

##The __str__ method returns a string representation of an object that is human-readable
  def __Str__(self):
      return self.value+' of '+(self.suit)

##__repr__ method returns a string representation of an object that is machine-readable

  def __repr__(self):
     return f'{self.value} of {self.suit}'
    

```

```python
##-------------------from resource------------------------##
class Deck:


    def __init__(self):


      self.cards = []
      for i in range (2,15):
          for j in range (4):
                self.cards\
                    .append(Card(i,
                                 j))
      shuffle(self.cards)

    def rm_card(self):
        if len(self.cards) == 0:
            raise Exception( 'Deck finished'
            )
        
        return self.cards.pop()
    
    ## method that python interperets when printing a self-defined object
    def __Str__(self):
      v = self.cards[self.value] +\
          " of " + \
            self.cards[self.suit]
      return (v)
```

```python
##--------------by resource------------------##

class Player:
  def __init__(self, name):
    self.name = name
    self.card =""
    self.wins_round = 0
    

  def __str__(self):
        return 'player '+self.name+' has won '+str(self.wins_round) +' rounds'
```

```python
##------------by Dana----------------------##

class Game:

  def __init__(self , num_players , players, deck):
    self.max_round = int(len(deck.cards)/num_players) 
    self.num_players = num_players
    self.players=players
    self.deck = deck
    self.round = 0 

  def switch_turn(self, current_player):
    length = len(self.players)
    for i in range (length):
      #if current player is last player: 1- check for max round 2- new round start from player 1
      if self.players[length-1] == current_player:
        if self.round == self.max_round:
          return current_player 
        else:
          self.round = self.round +1
          print('-------------------')
          current_player=self.players[0]
          return current_player 

      # if current player not last, go to next player
      elif self.players[i] == current_player:
        current_player = self.players[i+1]
        return current_player
  

  def draw_card(self ,current_player):
    card = self.deck.rm_card()
    current_player.card = card
    #score = card.score
    return card

#winner of round
  def round_winner(self):
    max = self.players[0]
    for i in range (len(self.players)-1):
      if(self.players[i+1].card > self.players[i].card):
        max = self.players[i+1] 
    max.wins_round+=1
    return max 

        
#winner of game
  def highest_score(self):
   highest = self.players[0]
   for i in range (len(self.players)-1):
     if self.players[i+1].wins_round > self.players[i].wins_round:
       highest = self.players[i+1]
   return highest

  def print_scores(self):

    print(f'\n {self.highest_score().name} has won the game! \n')
    print("-------Results------")
    for i in range (len(self.players)):
      print(f"score of {self.players[i].name} is {self.players[i].wins_round} winning rounds")




  #play a round
  def play_game(self, current_player):

    while True:
      #try:
      self.draw_card(current_player)
      print(current_player.name +" drew " + str(current_player.card))
      #check if a new round is starting
      if(self.switch_turn(current_player)== self.players[0]):
          # check if deck is finished
        if (self.round == self.max_round):
         break
        else:
          round_winner = self.round_winner()
          print(f'\n {round_winner.name} has won the round \n')
          print('-------------------')
          #print(f'round is {self.round}')
          #print(f'max round {self.max_round}')
          break
      #switch players for taking turns
      else:
        current_player = self.switch_turn(current_player)
      









#----------------------------by dana-----------------------#

def main():
  
  players=[]

  num_players=int(input("How many players?:"))
  print("\n")
  
  for i in range (num_players):
    #print(i ,num_players)
    print(f"Player {i+1}:")
    players.append(Player(input("Enter your name: ")))
    print("\n")



  #print_scoreBoard(score_board)

    #Start the game
    deck = Deck()
    game=Game(num_players, players, deck)
    current_player = players[0]

   #rounds looping
  while True:

    #Player choice menu
    print("Turn to play!")
    print("Enter 1 to play a round")
    print("Enter 2 to Quit")


    # Try block for handling user choices
    try:
      choice = int(input())
    except ValueError:
      print("Incorrect input! please choose from the list")
      continue

    if choice == 1:
      round_winner = game.play_game(current_player)
      if (game.round == game.max_round):
        print(f'\n Game Over!')
        winner = game.highest_score()
    
        print(f' \n player {winner.name} has won the game with {winner.wins_round} rounds \n ')
        game.print_scores()
        break
      continue

  

    elif choice ==2:
      print("#--------Final Scores---------#")
      game.print_scores()
      break
    else:
      print("incorrect choice, try again\n")






      
 
```
