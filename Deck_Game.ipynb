```python
from random import shuffle

##-----------------from resource-----------------------##
class Card:

  ## suits&value are represented by int

  suits = ["spades",
             "hearts",
             "diamonds",
             "clubs"]
    
  values = [None, None,"2", "3",
              "4", "5", "6", "7",
              "8", "9", "10",
              "Jack", "Queen",
              "King", "Ace"]

  def __init__(self, value,suit):
      self.value = value
      self.suit = suit

##  magic functions that enables python to compare self defined objects
      ##less than##
  def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.suit < c2.suit:
                return True
            else:
                return False
        return False
      ##greater than##
  def __gt__(self, c2):
        if self.value > c2.value:
            return True
        if self.value == c2.value:
            if self.suit > c2.suit:
                return True
            else:
                return False
        return False

## method that python interperets when printing a self-defined object
  def __Str__(self):
      return 'card='+self.value+' shape'+(self.suit)

  #def __repr__

```

```python
##-------------------from resource------------------------##
class Deck:
    suits = ["spades",
             "hearts",
             "diamonds",
             "clubs"]
    
    values = [None, None,"2", "3",
              "4", "5", "6", "7",
              "8", "9", "10",
              "Jack", "Queen",
              "King", "Ace"]

    def __init__(self):
      suits = ["spades",
             "hearts",
             "diamonds",
             "clubs"]
    
      values = [None, None,"2", "3",
              "4", "5", "6", "7",
              "8", "9", "10",
              "Jack", "Queen",
              "King", "Ace"]

      self.cards = []
      for i in range(2, 15):
          for j in range(4):
                self.cards\
                    .append(Card(values[i],
                                 suits[j]))
      shuffle(self.cards)

    def rm_card(self):
        if len(self.cards) == 0:
            return
        return self.cards.pop()
    
    ## method that python interperets when printing a self-defined object
    def __Str__(self):
      v = self.cards[self.value] +\
          " of " + \
            self.cards[self.suit]
      return (v)
```
